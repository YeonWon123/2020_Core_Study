* time 모듈의 사용

```
# 연습문제 7.4 - 의사난수 함수 만들기
# (1)
def my_rand(x):
  new_x = (28482331 * x + 59483) % 1000001
  return new_x

x = 59483
for _ in range(10): # 반복하는 변수를 for문 안에서 쓰지 않는 경우, _로 생략 가능
  x = my_rand(x)
  print(x)
```

```
# (2)
import time

def my_rand(x):
  new_x = (28482331 * x + 59483) % 1000001
  return new_x

seed = time.time()
x = int(seed) # int로 형변환해주는 것 주의!

for _ in range(5):
  x = my_rand(x)
  print(x)
```

* math 모듈의 사용

* 소수점 아래 x번째라리까지 나타내는 방법
 - '%0개수.자릿수f' % 숫자
 - '{인덱스:0개수.자릿수f'}.format(숫자)


```
# 연습문제 7.5
import math

for i in range(0, 180, 1):
  radian = math.radians(i)
  print("sin({0}) = {1:.3f} cos({2}) = {3:.3f}, tan({4}) = {5:.3f}".format(i, round(math.sin(radian), 3), i, round(math.cos(radian), 3), i, round(math.tan(radian), 3))) # 괄호
```

```
# 연습문제 7.6
import math

for i in range(1, 11, 1):
  print("sqrt({}) = {:.3f}".format(i, math.sqrt(i))) # {}에서 인덱스는 생략할 수 있음
```
